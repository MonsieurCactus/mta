```python
sta = 'R190'

oneStation = [d for d in data if d[1]==sta]
#oneStation = [d[3:] for d in oneStation]
[x[:3] for x in oneStation]
X = {}
for x in oneStation:
    if x[2] not in X.keys():
        X[x[2]] = x[3:]
    else:
        X[x[2]] += x[3:]

k = X.keys()[3]
X[k]

T =  np.array([24*int(X[k][5*t].split('-')[1])  + int(X[k][5*t+1].split(':')[0]) for t in range(len(X[k])/5) ])
T = T - T[0]

dt = 0.1
Tnew = np.arange(0, 160, dt)
In = np.interp(Tnew, T, np.array([ int(X[k][5*t+3]) for t in range(len(X[k])/5) ]))
Out = np.interp(Tnew, T, [ int(X[k][5*t+4]) for t in range(len(X[k])/5) ])

# http://web.mta.info/developers/resources/line_colors.htm
plt.plot(Tnew[1:]+12, np.convolve(In-Out, [1,-1])[1:-1]/dt, linewidth="2", color = "#EE352E")
plt.ylabel("ridership", fontsize=20)
plt.title("23st Station -- 1 Line", fontsize=20)
plt.xticks( np.arange(0, 160, 24)+12, ["Sat", "Sun", "Mon", "Tue", "Wed", "Thu", "Fri"])
plt.show()
```
