```python
def f(sta):
    # AGGREGATE
    #sta = 'R195'
    oneStation = [d for d in data if d[1]==sta]

    X = {}
    for x in oneStation:
        if x[2] not in X.keys():
            X[x[2]] = x[3:]
        else:
            X[x[2]] += x[3:]

    # ANALYZE + CLEAN  

    enter = np.zeros(1680)
    exit = np.zeros(1680)

    for k in X.keys():

        T =  np.array([24*int(X[k][5*t].split('-')[1])  + int(X[k][5*t+1].split(':')[0]) for t in range(len(X[k])/5) ])
        T = T - T[0]

        In = np.array([ int(X[k][5*t+3]) for t in range(len(X[k])/5) ])
        Out =  np.array([ int(X[k][5*t+4]) for t in range(len(X[k])/5) ])

        dt = 0.1
        Tnew = np.arange(0, 168, dt)
        In = np.interp(Tnew, T, In)
        Out = np.interp(Tnew, T, Out)

        T2 = Tnew[::10*4]
        In = In[::10*4]
        Out = Out[::10*4]

        dt = 0.1
        Tnew = np.arange(0, 168, dt)
        In = np.interp(Tnew, T2, In)
        Out = np.interp(Tnew, T2, Out)

        enter += In
        exit += Out
    return {"in": enter, "out": exit}

def graph(x, sta):
    # http://web.mta.info/developers/resources/line_colors.htm
    line = random.choice(remoteToStations[sta][1])
    plt.plot(Tnew[1:]+12, np.convolve(x, [1,-1])[1:-1]/dt, linewidth="3", color = color[line])
    plt.title("%s -- %s Line"%(remoteToStations[sta][0].title(),line), fontsize=20)
    plt.xticks( np.arange(0, 168, 24)+12, ["Sat", "Sun", "Mon", "Tue", "Wed", "Thu", "Fri"])
    plt.grid(True)
```
